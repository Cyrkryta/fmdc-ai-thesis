### Wilcoxon

```python
w_stat_corr, p_val_corr = wilcoxon(corr_d, corr_u)
print(f"Wilcoxon Test Corr: W={w_stat_corr:.3f}, p={p_val_corr:.3f}")
w_stat_mse, p_val_mse = wilcoxon(mse_d, mse_u)
print(f"Wilcoxon Test MSE: W={w_stat_mse:.3f}, p={p_val_mse:.3f}")
```

```python
def cohen_d(x, y):
    return (np.mean(y) - np.mean(x)) / np.sqrt((np.std(x, ddof=1) ** 2 + np.std(y, ddof=1) ** 2) / 2)

effect_size_corr = cohen_d(corr_d, corr_u)
effect_size_mse = cohen_d(mse_d, mse_u)
print(f"Cohen's d Effect Size Corr: {effect_size_corr:.3f}")
print(f"Cohen's d Effect Size MSE: {effect_size_mse:.3f}")
```

### T-test

```python
t_stat_corr, p_val_corr = ttest_rel(corr_d, corr_u)
t_stat_mse, p_val_mse = ttest_rel(mse_d, mse_u)
print(f"Paired t-test corr: t={t_stat_corr:.3f}, p={p_val_corr:.3f}")
print(f"Paired t-test mse: t={t_stat_mse:.3f}, p={p_val_mse:.3f}")
```
