```python
import torch
import os
from project.models.unet3d_fieldmap import UNet3DFieldmap
from project.data.data_util import normalize_img, nii2torch
import json
import tempfile
import subprocess
import nibabel as nib

device = "cpu"

def predict_fieldmap(model, input_tensor):
    print("Retrieving the fieldmap")
    with torch.no_grad():
        pred = model(input_tensor)[0].detach().cpu().numpy()
    print(f"Pred type: {type(pred)}, dimension: {pred.shape}")
    fmap_pred = pred[0]
    print(f"Final fieldmap shape: {fmap_pred.shape}")
    fmap_nifti = np.transpose(fmap_pred, axes=(1,2,0))
    print(f"Final fieldmap shape (Nifti): {fmap_nifti.shape}")
    return fmap_nifti

def apply_fugue_correction(bold_4d, pred_fieldmap, affine, echo_spacing, unwarp_direction, out_path):
    custom_tmp = "/indirect/student/magnuschristensen/dev/fmdc/fmdc-ai-thesis/helper-scripts/kfcv/tmpdir"
    print(echo_spacing, unwarp_direction)
    with tempfile.TemporaryDirectory(dir=custom_tmp) as tmpdir:
        bold_path = os.path.join(tmpdir, "bold.nii.gz")
        fmap_path = os.path.join(tmpdir, "fmap.nii.gz")
        print(bold_4d.shape, pred_fieldmap.shape)
        nib.save(nib.Nifti1Image(bold_4d, affine), bold_path)
        nib.save(nib.Nifti1Image(pred_fieldmap, affine), fmap_path)

        # Build the FUGUE command
        cmd = [
            "fugue",
            "-i", bold_path,
            "--loadfmap=" + fmap_path,
            "--dwell=" + str(echo_spacing),
            "--smooth3=3",
            "--unwarpdir=" + unwarp_direction,
            "-u", out_path
        ]

        subprocess.run(cmd, check=True)

def prepare_model_tensor(bold_path, t1w_path):
    # Load the images
    bold = nib.load(bold_path).get_fdata()
    t1w = nib.load(t1w_path).get_fdata()
    # Rep and repititions
    num_timesteps = bold.shape[3] if bold.ndim == 4 else 1
    if len(t1w.shape) == 3:
        t1w = np.repeat(t1w[None, :], num_timesteps, axis=0)
        t1w = np.transpose(t1w, axes=(1, 2, 3, 0))
    bold = nii2torch(bold)
    t1w = nii2torch(t1w)
    # Normalization
    t1w = normalize_img(t1w, 150, 0, 1, -1)
    max_bold_img = np.percentile(bold, 99)
    min_bold_img = 0
    bold = normalize_img(bold, max_bold_img, min_bold_img, 1, -1)
    bold_affine = nib.load(bold_path).affine
    # Convert to original 4d (H, W, D)
    bold_4d = np.transpose(bold, axes=(1, 2, 3, 0))
    t1w_4d = np.transpose(t1w, axes=(1, 2, 3, 0))
    # Retrieve mid index and slice
    mid_idx = bold_4d.shape[3] // 2
    bold_ref = bold_4d[:, :, :, mid_idx]
    t1w_ref = t1w_4d[:, :, :, mid_idx]
    # Construct stacked input image (Numpy, Tensor)
    input_img = np.stack((bold_ref, t1w_ref))
    print(f"Input image shape: {input_img.shape}")
    input_tensor = torch.from_numpy(input_img).unsqueeze(0).float().to(device)
    print(f"Input tensor shape: {input_tensor.shape}")
    # Return the BOLD and T1w image
    return input_tensor, bold_affine
    
```
