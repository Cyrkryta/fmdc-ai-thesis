```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data
val_loss_path = "/indirect/student/magnuschristensen/dev/fmdc/helper-scripts/Visualization/bs8_vl1300_300325_valloss.csv"
train_loss_path = "/indirect/student/magnuschristensen/dev/fmdc/helper-scripts/Visualization/bs8_tl450_300325_trainloss.csv"
val_loss = pd.read_csv(val_loss_path)
train_loss = pd.read_csv(train_loss_path)
val_loss_vals = val_loss["elated-water-90 - val_loss"]
train_loss_vals = train_loss["elated-water-90 - train_loss"]
epochs = train_loss["epoch"]
```

```python
train_loss_vals
```

```python
plt.figure()
plt.plot(train_loss_vals, label="Training loss")
plt.plot(val_loss_vals, label="Validation loss")
plt.xlabel("Epoch")
plt.ylabel("Mean-Squared Error (MSE)")
plt.title("Batch Size 8, default params\n2700 train, 300 validation")
plt.legend()
plt.annotate(text=round(train_loss_vals[len(train_loss_vals)-1], 2), xy=(epochs[len(epochs)-1], train_loss_vals[len(train_loss_vals)-1]))
plt.annotate(text=round(val_loss_vals[len(val_loss_vals)-1], 2), xy=(epochs[len(epochs)-1], val_loss_vals[len(val_loss_vals)-1]))
plt.grid(True)
```
